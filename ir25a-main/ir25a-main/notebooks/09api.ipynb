# 2.Retrieval

### 2.1 Cargo el corpus 20 news groups

```python
from sklearn.datasets import fetch_20newsgroups

newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
newsgroupsdocs = newsgroups.data
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
import pandas as pd
```

```python
newsgroups_df = pd.DataFrame(newsgroupsdocs[:1000], columns=['raw'])
newsgroups_df
```

### 2.2 Transformo a embeddings

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(newsgroupsdocs)
```

```python
embeddings
```

```python
newsgroups_df['embedding'] = list(embeddings[:1000])
newsgroups_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
query = "Russian Federation"
query_emb = model.encode(query)
print(query_emb)
```

```python
similarities = cosine_similarity(query_emb.reshape(1, -1), embeddings[:1000])
```

Obtengo los 5 documentos mas similares a mi query

```python
z = zip(newsgroups_df.index, similarities)
print(sorted(z, key=lambda x: x[1], reverse=True)[:5])
```
